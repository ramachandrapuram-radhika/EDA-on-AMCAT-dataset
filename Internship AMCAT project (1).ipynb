## Univariate Analysis

```python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("*"*10, col_name, "*"*10)
        print(numerical_data[col_name].agg(['min','max','mean','median','std','skew','kurt']))
        print()
```

# Research Questions

### 1. “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.


```python
grouped_df = df.groupby(['Designation', 'Specialization']).agg({'Salary': 'mean'}).reset_index()

# Checking if the average salaries are within the range of 2.5-3 lakhs

lower_bound = 250000
upper_bound = 300000
result = (df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)
print(grouped_df)
if result.all():
    print("The claim is correct for the given data.")
else:
    print("The claim is not correct for the given data.")
```
